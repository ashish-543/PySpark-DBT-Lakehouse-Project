#### Creating new catalog

```python
spark.sql("CREATE CATALOG IF NOT EXISTS lakehouse")
```

#### Using the new catalog

```python
spark.sql("USE CATALOG lakehouse")
```

#### Creating schemas

```python
spark.sql("""
          CREATE SCHEMA IF NOT EXISTS source
          COMMENT 'Source: raw data'
          """)
spark.sql("""
          CREATE SCHEMA IF NOT EXISTS bronze
          COMMENT 'Bronze layer: raw ingested data'
          """)
spark.sql("""
          CREATE SCHEMA IF NOT EXISTS silver
          COMMENT 'Silver layer: cleansed data'
          """)
spark.sql("""
          CREATE SCHEMA IF NOT EXISTS gold
          COMMENT 'Gold layer: business ready data'
          """)
```

#### Creating volume inside source schema

```python
spark.sql("""
          CREATE VOLUME IF NOT EXISTS lakehouse.source.raw_data
          COMMENT 'Volume that contains raw CSV files'
          """)
```
